```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display as display
import glob as glob
```

```python
y, sr = librosa.load("../music/divertimento.m4a", offset=60,duration=20)
fig, ax = plt.subplots(figsize=(16,4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
img = display.specshow(D, x_axis='time',
                               sr=sr, ax=ax,y_axis='fft_note')
ax.set(title='Divertimento')
```

```python
diana = glob.glob('../music/*.wav')
for archivo in diana:
    y, sr = librosa.load(archivo)
    fig, ax = plt.subplots(figsize=(12,12))
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    img = display.specshow(D, x_axis='time',
                                   sr=sr, ax=ax,cmap='hot',y_axis='fft_note')
    # ax.set_ylim(0,2000)
    ax.set(title=archivo)
```
