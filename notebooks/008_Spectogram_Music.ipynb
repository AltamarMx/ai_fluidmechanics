```python
y, sr = librosa.load("../music/divertimento.m4a", offset=60,duration=20)
fig, ax = plt.subplots(figsize=(16,4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
img = display.specshow(D, y_axis='linear', x_axis='time',
                               sr=sr, ax=ax)
ax.set(title='Divertimento')
```
