```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.transform import Rotation as R
```

```python
p = np.array([0,1,0])
```

```python
r = R.from_euler("x",180,degrees=True)
r.as_matrix().round(2)

pr = r.apply(p)
pr.round(2)
r.as_matrix().round(2)
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)

fig.add_axes(ax)
ax.scatter(0,0,0,color="k",s=80)
ax.scatter(p[0],p[1],p[2],color="b",s=80)
ax.scatter(pr[0],pr[1],pr[2],color="red",s=80)

# ax.quiver3D(0,0,0,pr[0],pr[1],pr[2] ,color="green")
decora()

# ax.quiver3D(0,0,0,pr[0],pr[1],pr[2] ,color="green")
```

```python
def decora():
    ax.set_xlabel('$x_1$ (x)')
    ax.set_ylabel('$x_2$ (y)')
    ax.set_zlabel('$x_3$ (z)')


    ax.quiver3D([0],[0],[0],[1],[0],[0] ,color="green",label="x")
    ax.quiver3D([0],[0],[0],[-1],[0],[0],color='g', linestyle="dashed")
    ax.quiver3D([0],[0],[0],[0],[1],[0],color="blue",label="y")
    ax.quiver3D([0],[0],[0],[0],[-1],[0],color="blue",linestyle="dashed")
    ax.quiver3D([0],[0],[0],[0],[0],[1],color="red",label="z")
    ax.quiver3D([0],[0],[0],[0],[0],[-1],color="red",linestyle="dashed")

    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
    ax.set_zlim(-1,1)
    ax.legend()
```
