```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
x, z = make_circles(n_samples=200,noise=0.05, factor=0.2)
```

```python
plt.scatter(x[:,0],x[:,1],c=z)
plt.axis("equal")
```

```python
x_train = x
y_train = z
model = keras.Sequential()

model.add(keras.Input(shape=8))
model.add(layers.Dense(1,activation='relu'))
# model.add(layers.Dense(1,activation='sigmoid'))
model.compile(loss='MSE',optimizer=keras.optimizers.SGD(learning_rate=.01))

problem = model.fit(x_train,y_train,epochs=1000,verbose=False)
```

```python
plt.plot(problem.history['loss'])
```
